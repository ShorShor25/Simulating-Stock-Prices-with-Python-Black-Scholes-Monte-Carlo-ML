```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from SIM.GBM import simulate_GBM
```

```python
S0 = 100
mu = 0.05
sigma = 0.2
T = 1.0
N = 500
```

```python

time, prices = simulate_GBM(S0, mu, sigma, T, N)

df = pd.DataFrame({'Time': time, 'Price': prices})
df.to_csv('market_data.csv', index=False)

print("Market data saved to 'market_data.csv'.")
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['Time'], df['Price'])
plt.title("Simulated Market History ('Real' Path)")
plt.xlabel("Time (Years)")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```
